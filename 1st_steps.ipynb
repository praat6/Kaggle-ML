```python
#la observació número 506 es un nan
data1 = data.drop([506]).reset_index(drop=True) #reset index per poder fer el següent for
tweets_list = data1["text"]
print(tweets_list.head())
data1.head()
```

```python
l = []
for i in range(0,len(tweets_list)):
    tweet = tweets_list[i]
    l.append(tweet.split(sep=" "))

len(l) #l es una llista de llistes amb cada 1 dels tweets
```

```python
import re #regular expressions

tags=[]
for i in range(0, len(l)):
    iterable_l = l[i]
    for j in range(0,len(iterable_l)):
        i_j = iterable_l[j]
        m = re.search(r"@\w+", i_j)
        if m!=None:
            tags.append(m.group())
        else:
            pass
tags #treure tota la gent etiquetada
```

```python
import numpy as np

print("en total hi ha " + str(len(tags)) + " tags (contant repetits) dins els 2880 tweets")
print("en total hi ha " + str(len(np.unique(tags)))+ " tags diferents")
```
